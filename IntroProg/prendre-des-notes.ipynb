# Prendre des notes

L'air de rien, dans les quelques feuilles sur lesquelles vous avez travaillé, vous avez
déjà rencontré un bon nombre de noms, de mots clés, de symboles, de concepts: `int`,
`variable`, `expression conditionnelle`, `%` ...

Comme toute science, **l'informatique a son vocabulaire et il est important de le
maîtriser avec précision pour pouvoir comprendre et communiquer efficacement**. Une
stratégie efficace pour mémoriser ce vocabulaire et ces définitions est de se faire des
fiches de synthèse. Cela est particulièrement vrai dans un cours comme celui-ci où les
éléments sont introduits au fil de l'eau.

:::{admonition} Prenez des notes
:class: tip

1. Sur une feuille de papier (ou tablette, ...), notez de mémoire tous les mots,
   symboles, concepts que vous avez déjà croisé dans ce cours. Vous pouvez compléter
   chacun de ces éléments avec quelques mots pour vous rappeler la définition, ou pour
   indiquer dans quelle feuille chercher pour retrouver les détails.

2. Reparcourez rapidement les feuilles sur lesquelles vous avez travaillé afin de repérer
   tous les éléments que vous n'aviez pas notés.

3. Sur une nouvelle feuille de papier, réorganisez visuellement votre fiche de synthèse
   pour structurer les éléments. Vous pouvez par exemple faire une carte mentale; ou bien
   regrouper les éléments par catégories (exemple; types, opérateurs, concepts, ...).

4. Par la suite, complétez au fur et à mesure votre fiche de synthèse avec les nouveaux
   éléments que vous rencontrerez. Réorganisez la régulièrement votre fiche de synthèse
   pour garder une vue d'ensemble.
:::

:::{admonition} Astuce
:class: hint

Vous pouvez réaliser de telles synthèses à plusieurs autour d'un tableau ou d'une grande
feuille, en utilisant des post-its pour pouvoir facilement déplacer et réorganiser les
éléments.
:::
