### Les instructions conditionnelles

Nous allons maintenant introduire formellement les instructions conditionnelles, et tout
d'abord spécifier comment regrouper plusieurs instructions ensemble comme ci-dessus (et
comme apperçu précédemment avec les fonctions.; cela se fait en Python à l'aide de
l'indentation :

(bloc_d_instruction)=
:::{admonition} Définition : Bloc d'instructions
Un ***bloc d'instructions*** est une suite d'une ou plusieurs instructions à
exécuter successivement. Il est délimité par son **indentation** (par convention quatre
espaces de plus que le code environnant) ainsi que par un ':' sur la ligne qui précède :

```python
    ...:

        instruction 1
        instruction 2
        ...
        instruction n

    ...
```
:::

Nous pouvons maintenant définir les instructions conditionnelles simples et alternatives.


(instruction_conditionnelle_simple)=
::::{admonition} Définition : Instruction conditionnelle simple : «si ... alors ...»
:::{admonition} Syntaxe :
```python
if condition:
    bloc d instructions
```
:::

:::{admonition} Sémantique
1. Évaluation de la condition
2. Si sa valeur est vraie, exécution du bloc d'instructions
:::
::::

:::{admonition} Exemple
```python
if regarde() == Toile:     # Au secours, fuyons!
    gauche()
    gauche()
```
:::

(instruction_conditionnelle_alternative)=
::::{admonition} Définition : Instruction conditionnelle alternative : «si ... alors ... sinon ...»
:::{admonition} Syntaxe
```python
if condition:
    bloc d instructions 1
else:
    bloc d instructions 2
```
:::

:::{admonition} Sémantique
1. Évaluation de la condition
2. Si sa valeur est «Vrai», exécution du bloc d'instructions 1
3. Si sa valeur est «Faux», exécution du bloc d'instructions 2
:::
::::

:::{admonition} Exemple
```python
if regarde() == Toile:    # Au secours, fuyons!
    gauche()
    gauche()
else:                     # Tout va bien
    avance()
```
:::

(instruction_conditionnelle_filee)=
::::{admonition} Définition : Instruction conditionnelle filée : «si ... alors ... sinon si ... alors sinon ...»
:::{admonition} Syntaxe
```python
if condition1:
    bloc d instructions 1
elif condition2:
    bloc d instructions 2
else:
    bloc d instructions 3
```
:::

:::{admonition} Sémantique
Le code ci-dessus est strictement équivalent à (mais plus lisible que) :

```python
if condition1:
    bloc d instructions 1
else:
    if condition2:
        bloc d instructions 2
    else:
        bloc d instructions 3
```
:::
::::

Il peut y avoir autant de clauses `elif` que nécessaire.
