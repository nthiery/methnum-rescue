# Prise en main des carnets Jupyter

Un ***carnet interactif*** est un document interactif permettant de mixer narration,
interaction, visualisation, calcul ou programmation. Dans ce premier carnet, vous allez
découvrir les trois opérations essentielles pour utiliser un carnet: lire, exécuter,
éditer.

:::{admonition} Définition
Un carnet interactif est composé d'une séquence de ***cellules***. Chaque cellule peut
contenir du texte riche, un calcul à effectuer, des visualisations, etc. La ***cellule
active*** est encadrée, ce qui permet de voir sa portée. C'est l'auteur du document qui a
préalablement spécifié les propriétés de chaque cellule.
:::
