# Enrichir les zones de textes (Maths, médias, liens)

:::{admonition} Objectif pédagogique
Découvrir la syntaxe permettant d'inclure des formules mathématiques, des images ou
vidéos et liens dans les zones de texte Markdown
:::

Les carnets Jupyter permettent de présenter de manière claire et ordonnée des
informations. Pour cela vous pouvez enrichir vos textes :

- Avec des formules mathématiques
- Avec des médias (images ou vidéos)
- Avec des hyperliens

<div class="alert alert-info">

**La syntaxe :** (attention afin de voir la syntaxe, entrez en mode édition)

Pour écrire une formule :

- Utilisez <kbd>\$\$</kbd> pour entourer votre formule mathématique (un peu le même
  principe qu'en LaTeX).

Exemple : \$$\frac 1 {1-\frac 1z} = \sum_{i=0}^\infty \frac 1 {z^i}$\$

<!-- Ajouter une liste des opérations et symboles \frac \ sum --> 

Pour insérer un média :

- <img src="media/blackhole.jpg" alt="un trou noir" width="300"/>

Pour insérer un hyperlien :

- \[Texte du lien\](Url visée)

</div>

:::{admonition} Exercice
Ajoutez des cellules avec <kbd>A</kbd> en mode commande et exercez-vous à insérer des
formules, des images ou vidéos et des hyperliens.
:::

## Bilan

Rendre le plus lisible possible un carnet vous aidera à mieux organiser votre réflexion,
à pouvoir collaborer plus facilement et éventuellement préparer des fiches ou vos
révisions.
