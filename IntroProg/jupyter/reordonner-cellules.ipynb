```python
# Étape 2
def unarrondi(orientation,avancer):
    if orientation == "L":
        for i in range (0,45):
            t.left(2)
            t.forward(1)
    else:
        for i in range (0,45):
            t.right(2)
            t.forward(1)
    if avancer=="V":
        t.forward(60)
```

```python
# Étape 7
t.forward(15)
unarrondi("R","V")
unarrondi("R","F")
t.forward(70)
t.left(90)
t.forward(12)
t.left(90)
t.forward(45)
t.right(90)
t.forward(20)
unarrondi("R","V")
t.forward(2)
unarrondi("R","F")
t.forward(45)
unarrondi("R","V")
t.forward(10)
unarrondi("L","F")
t.forward(45)
```
