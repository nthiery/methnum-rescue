# Rechercher et remplacer

:::{admonition} Objectif pédagogique
Savoir rechercher et remplacer des termes dans un carnet
:::

::::{admonition} Exercice
Parmi les variables ci-dessous, certaines ont des noms imcompréhensibles. Recherchez et
remplacez le nom des variables pour que le script prenne plus de sens. Le script affiche
des tables de multiplication. Vous pouvez l'éxecuter pour voir le résultat.<br>

:::{admonition} Indication
:class: tip

Basculez en mode commande ([revoir le tuto mode commande](mode-edition-vs-commande.ipynb))
et utilisez `F` pour rechercher et remplacer.
:::
::::

```python
TOTO=list(range(1,5))
TOTO2=list(range(1,10))
for bipblop in TOTO:
    print("table de ",TOTO[bipblop-1], " :")
    for secondo in TOTO2: 
        print(str(TOTO[bipblop-1])," x ",str(TOTO2[secondo-1])," = ",(TOTO[bipblop-1]*TOTO2[secondo-1]))
    print("----------------")
```

## Bilan

La fonction rechercher et remplacer vous permet de changer facilement le nom d'une
variable dans tous le carnet sans la modifier à la main une par une. Cela évite des
erreurs de frappes, et des oublis.
