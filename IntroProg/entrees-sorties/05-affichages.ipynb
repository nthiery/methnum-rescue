# Affichages

Dans les prochaines feuilles, nous aurons besoin d'afficher des informations pour tracer
pas à pas l'exécution d'un programme. Nous vous donnons ici le minimum nécessaire pour
cela. Nous reviendrons dessus plus en détail ultérieurement.

L'affichage à l'écran est une des façons de communiquer entre le programme et
l'utilisateur. Comme on l'a vu jusqu'à présent, lorsque l'on exécute une cellule dans
Jupyter, la valeur de la *dernière expression* de la cellule est affichée :

```python
1+2
3+4
5+6
```

Les autres calculs ont bien été effectués par la machine, mais sans montrer les valeurs
obtenues.

Nous allons maintenant voir comment afficher des informations. La cellule suivante
affiche le résultat du calcul :

```python
print(1+2)
```

On peut afficher plusieurs valeurs à la suite :

```python
print(1+2)
print(3+4)
print(5+6)
```

On peut aussi afficher du texte (chaînes de caractères) en utilisant les guillemets :

```python
print("Bonjour, comment allez-vous ?")
```

On peut afficher la valeur d'une expression obtenue par évaluation d'une combinaison
d'opérateurs et de variables :

```python
a = 1
print(a + 2)
```

On peut passer plusieurs arguments à la fonction print en les séparant par une virgule,
`print` les affichera par **concaténation** en utilisant un espace vide par défaut comme
séparateur.

(concatenation)=
:::{admonition} Définition : Concaténation
En programmation, on appelle la ***concaténation*** de deux chaînes de
caractères la chaîne formée de ces deux chaînes mises bout à bout. Le terme
***concaténation*** désigne également l'opération de concaténer des chaînes.
:::

```python
age = 32
print("J'ai ", age, " ans.") 
```
