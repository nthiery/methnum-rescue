:::{admonition} Consigne

Écrire une fonction `PGCD` qui prend en entrée deux entiers `a` et `b` non nuls et renvoie leur [Plus Grand Diviseur Commun](https://fr.wikipedia.org/wiki/Plus_grand_commun_diviseur_de_nombres_entiers).

:::

```python
def PGCD(a, b): 
  
    # si b=0 
    if (b==0): 
        # renvoi de a 
        return a 
    else: # sinon 
        # calcul du reste r de la division de a par b 
        r = a % b 
  
        # appel de la fonction PGCD pour a=b et b=r 
        return PGCD(b, r)
```

```python
assert PGCD(221, 782) == 17, f"Valeur attendue : 17, valeur obtenue : {PGCD(221, 782)}"
assert PGCD(1, 2) == 1, f"Valeur attendue : 1, valeur obtenue : {PGCD(1, 2)}"
assert PGCD(10, 14) == 2, f"Valeur attendue : 2, valeur obtenue : {PGCD(10, 14)}"
assert PGCD(40, 27) == 1, f"Valeur attendue : 1, valeur obtenue : {PGCD(40, 27)}"
```
