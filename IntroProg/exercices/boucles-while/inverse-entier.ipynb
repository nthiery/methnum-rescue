Définir une fonction `retourne` qui prend en paramètre un entier `n`
et retourne les chiffres de ce nombre. Par exemple, `2134` donnera
`4312`.

:::{admonition} Indications
Pour extraire successivement les chiffres, vous pourrez utiliser les
opérateurs modulo `a % b`, qui donne le reste de la division
euclidienne de `a` par `b`, et `a // b`, qui divise `a` par `b` et
renvoie le quotient entier de cette division sans tenir compte de la
partie décimale.
:::

::::

```python
def retourne(n):
    temp = n
    reverse = 0
    while n > 0:
        remainder = n % 10
        reverse = (reverse * 10) + remainder
        n = n // 10
    return reverse
```

```python
assert retourne(782) == 287, f"Valeur attendue pour retourne(782) : {287}, valeur obtenue : {retourne(782)}"
assert retourne(100) == 1, f"Valeur attendue pour retourne(100) : {1}, valeur obtenue : {retourne(100)}"
assert retourne(1234) == 4321, f"Valeur attendue pour retourne(1234) : {4321}, valeur obtenue : {retourne(1234)}"
assert retourne(12345) == 54321, f"Valeur attendue pour retourne(12345) : {54321}, valeur obtenue : {retourne(12345)}"
```
