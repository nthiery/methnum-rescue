```python
from random import randint
from jupylates.jupylates_helpers import SUBSTITUTE, INPUT_TEXT
SUBSTITUTE(
    N = randint(0, 20),
	T1 = randint(3, 8),
	T2 = randint(12, 17)
)
```

:::{admonition} Consigne
Quelle est la valeur de `S` après les instructions suivantes  :
:::

```python
X = N
if X <= T1:
    S = "A"
elif X <= T2:
    S = "B"
else:
    S = "C"
```

:::

```python
solution, answer = INPUT_TEXT("S", S)

assert solution == answer
```
