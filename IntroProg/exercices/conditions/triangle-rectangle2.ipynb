:::{admonition} Consigne

Définir une fonction `est_triangle_rectangle(a, b, c)` prenant en
entrée trois entiers `a`, `b` et `c` qui représentent les longueurs
des côtés d'un triangle, et renvoyant `True` si le triangle est
rectangle, `False` sinon.

**Indication :** Cette fois, l'hypoténuse peut être en `a`, `b` ou
`c`.

:::

```python
def est_triangle_rectangle(a, b, c):
    # Vérifier le théorème de Pythagore dans les trois cas
    return a**2 == b**2 + c**2 or \
           b**2 == a**2 + c**2 or \
           c**2 == a**2 + b**2
```

```python
assert est_triangle_rectangle(8, 10, 6), f"est_triangle_rectangle(8, 10, 6) devrait renvoyer True"
assert not est_triangle_rectangle(10, 7, 6), f"est_triangle_rectangle(10, 7, 6) devrait renvoyer False"
assert est_triangle_rectangle(16, 20, 12), f"est_triangle_rectangle(16, 20, 12) devrait renvoyer True"
assert not est_triangle_rectangle(200, 16, 12), f"est_triangle_rectangle(200, 16, 12) devrait renvoyer False"
```
