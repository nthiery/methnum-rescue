:::{admonition} Consigne

Définissez une fonction `distance` qui prend en entrée quatre
paramètres `xa`, `ya`, `xb` et `yb` représentant les cordonnées de
deux points $a$ et $b$ et qui renvoie leur distance relative
$\sqrt{(x_b-x_a)^2 + (y_b-y_a)^2}$. Vous disposez de la fonction
`math.sqrt(c)` qui renvoie la racine carrée du nombre `c`.

:::

```python
def distance(xa, ya, xb, yb):
    return math.sqrt((xb-xa)**2+(yb-ya)**2)
```

```python
def SOLUTION(xa, ya, xb, yb):
    return math.sqrt((xb-xa)**2+(yb-ya)**2)
```

```python
precision = 0.001
for parameters in [[0, 0, 0, 0], [1, 2, 3, 4], [10, -2, 34, 48], [55, 2000, -3, 0]]:
    assert abs(distance(*parameters) - SOLUTION(*parameters)) <= precision, f"Valeur attendue pour distance({parameters}) : {SOLUTION(*parameters)},  valeur obtenue : {distance(*parameters)}"
```
