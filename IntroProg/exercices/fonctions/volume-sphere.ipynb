:::{admonition} Consigne

Définissez une fonction `volume` qui prend en paramètre le rayon
`r` d'une boule et calcule son volume $\frac 43 \pi r^3$. Vous
disposez de la variable `math.pi` qui a pour valeur une bonne
approximation de $\pi$.

:::

```python
def volume(r):
    return 4 * math.pi * r**3 / 3
```

```python
import math

def SOLUTION(r):
    return 4 * math.pi * r**3 / 3

for r in [0, 3, 20, 1000]:
    assert abs( volume(r) - SOLUTION(r) ) <= 0.01,  f"Valeur attendue pour volume({r}) : {SOLUTION(r)}, valeur obtenue : {volume(r)}" 
```
