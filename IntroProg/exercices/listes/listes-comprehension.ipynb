````python
from jupylates.jupylates_helpers import SUBSTITUTE
import random

var = random.choice("abcijlkpq")
varT = random.choice("st")

seuil = random.randint(3, 5)
condition = f"{var} {random.choice(['<', '>'])} {seuil}"
expression = random.choice([var, f"{var}**2", f"2*{var}"])
COMPREHENSION = f"[{expression} for {var} in {varT} if {condition}]"

SUBSTITUTE(
    T=varT,
    COMPREHENSION=COMPREHENSION)
````
