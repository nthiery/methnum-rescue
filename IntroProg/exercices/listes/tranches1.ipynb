Objectifs pédagogique : Les tranches de listes

:::{admonition} Exercice

Construisez une liste `semaine` contenant les sept jours de la
    semaine. À partir de cette liste, utilisez deux tranches pour
    extraire respectivement les cinq premiers jours de la semaine
    d'une part, et ceux du week-end d'autre part.

:::

```python
### BEGIN SOLUTION
semaine = ["lundi", "mardi", "mercredi", "jeudi", "vendredi", "samedi", "dimanche"]
premiers_jours = semaine[:5]
weekend = semaine[5:]
### END SOLUTION
```

```python
assert semaine == ["lundi", "mardi", "mercredi", "jeudi", "vendredi", "samedi", "dimanche"], f"semaine devrait contenir ['lundi', 'mardi', 'mercredi', 'jeudi', 'vendredi', 'samedi', 'dimanche'] et contient {semaine}" 
assert premiers_jours == ["lundi", "mardi", "mercredi", "jeudi", "vendredi"], f"Les premiers jours devraient contenir ['lundi', 'mardi', 'mercredi', 'jeudi', 'vendredi'] et contiennent {premiers_jours}" 
assert weekend == ["samedi", "dimanche"], f"le weekend devrait contenir [ 'samedi', 'dimanche'] et contient {weekend}" 
```
