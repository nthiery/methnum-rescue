Utilisez une boucle `for` avec un compteur et un accumulateur. 

% Voir le cours Boucles/03-compteurs-accumulateurs.md

:::

::::

```python
def puissance(a, b):
    résultat = 1

    for i in range(b):
        résultat *= a   # Équivalent à résultat = résultat * a

    return résultat
```

```python
assert puissance(2, 3) == 2**3, f"Valeur attendue : {2**3}, valeur obtenue : {puissance(2, 3)}"
assert puissance(-3, 2) == (-3)**2, f"Valeur attendue : {(-3)**2}, valeur obtenue : {puissance(-3, 2)}"
```
