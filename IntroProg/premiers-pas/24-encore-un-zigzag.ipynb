# Encore un zigzag

Il faut parcourir ce labyrinthe en zigzag mais le nombre de lignes et de colonnes varie !

:::{attention}
Cet exercice est marqué d'un ♣ et est donc d'un niveau plus avancé ; n'hésitez pas à le
passer s'il vous paraît difficile afin de ne pas perdre trop de temps, et à revenir
dessus ultérieurement.
:::

```python
from laby.global_fr import *
from random import randint
n = randint(0,3)
m = randint(2, 20)
bloc_porte = "o " + "o " * m + "o\n"
bloc_porte+= "o " + ". " * m + "x\n"
bloc_porte+= "o . " + "o " * m + "\n"

bloc1 = "o " + ". " * m + "o\n"
bloc1+= "o " * m + ". o\n"

bloc2 = "o " + ". " * m + "o\n"
bloc2+= "o . " + "o " * m + "\n"

bloc_fourmi = "o " + "→ " + ". " * (m-1) + "o\n"
bloc_fourmi+= "o " + "o " * m + "o\n"

carte = bloc_porte + bloc1 + (bloc2 + bloc1) *n + bloc_fourmi
Laby(carte = carte)
```
