## Présentation du langage Python

Développé depuis $1989$. Principales caractéristiques :

-   **libre** et **gratuit**

-   **simple et très lisible**;

-   **une documentation** complète et simple disponible sur le web;

-   **utilisé massivement**, notamment pour le calcul scientifique, le
    traitement et la visualisation de données, l'apprentissage
    statistiques (IA);

    :::{figure} plots/popularite_langage.webp
	:width: 50%
	:::

-   **bibliothèque de base** très fournie;

-   très nombreuses **bibliothèques libres** développées par une
    communauté mondiale et bien testées dans tous les domaines;

-   **grande portabilité**: indépendant vis à vis du système
    d'exploitation (Linux, Windows, MacOS);

-   **orienté objet** avec **typage dynamique**;

-   présente un **support pour s'interfacer avec d'autres langages**.
