## Unix et GNU/Linux : Généralités

**Unix** est un système d'exploitation (ou OS pour operating system) :
il gère le fonctionnement et la répartition des différentes ressources,
que ce soit pour l'utilisateur ou les processus (dont les logiciels).\
Ces ressources sont aussi bien de la puissance de calcul (processeur ou
CPU) que le stockage (mémoire vive, disque dur, clef usb, etc.) et
les différents périphériques (clavier, souris, écran, enceintes \...)\

en sont dérivés : **MacOS**, **GNU/Linux**, **BSD**, **Android**.\
(**Windows** n'appartient pas à cette famille).\
**GNU/Linux** est de plus un OS libre. En particulier, chacun peut
l'utiliser, le modifier, le copier et le distribuer librement.

## Pour survivre dans un terminal : B.A. BA

Voici quelques commandes pour survivre dans un terminal en cas de
besoin:

-   [`pwd`]{style="color: red"} : *print working directory*, affiche
    l'adresse du répertoire courant.

-   [`ls`]{style="color: red"} : *list segments*, liste les fichiers et
    dossiers dans le répertoire courant.
    [`ls -lhrt`]{style="color: red"} pour plus d'information.

-   [`cd`]{style="color: red"} : *change directory*, change de
    répertoire courant. [`cd ..`]{style="color: red"} pour remonter
    l'arborescence, [`cd sous_repertoire`]{style="color: red"} pour
    descendre.

-   [`mkdir repertoire_cree`]{style="color: red"} : *make directory*,
    crée un répertoire dans le répertoire courant.

-   [`cp fichier1 fichier2`]{style="color: red"} : *copy*, copie le
    contenu de `fichier1` dans un fichier nommé `fichier2`.
    [`cp fichier1 dossier2/`]{style="color: red"} : copie/colle
    `fichier1` dans le répertoire `dossier2`.

-   [`mv fichier1 dossier2/`]{style="color: red"} : *move*, coupe/colle
    `fichier1` dans le répertoire `dossier2`.

-   [`rm fichier1`]{style="color: red"} : *remove*, supprime
    définitivement un fichier. **!!**

-   [`man nom_de_commande`]{style="color: red"} : accède au manuel
    d'aide linux pour la commande. Pour sortir, taper `q` (*quit*).

## Comment exécuter un code Python ?

De manière générale, il y a deux techniques principales pour traduire un
code source en langage machine:

-   la compilation: application tierce, appelée compilateur, transforme
    les lignes de code en langage machine dans un fichier exécutable.

-   l'interprétation: un interpréteur permet de traduire ligne par ligne
    le programme en langage machine.

**Python** : langage interprété qui fait appel à des modules compilés.\

-   **Mode programmation**:

    -   stocker le code (et son résultat) dans un **Jupyter Notebook**
        (fichier d'extension `.md` ou `.ipynb`), lu et édité via un
        navigateur internet (e.g. Firefox) (environnement convivial,
        facilitant les interactions) : **mode préféré dans ce cours**

    -   écrire le code à exécuter dans un fichier d'extension `.py`.
        L'exécuter dans un terminal avec la commande
        `python nomdufichier.py` ou `ipython nomdufichier.py`.

-   **Mode interpréteur:** dans un terminal, tapez `python` (classique)
    ou `ipython` (interpréteur plus évolué) pour accéder à un
    interpréteur Python. Tapez les instructions Python à exécuter.\

## Types des objets en Python

En python on trouve différents types de base pour les objets. Les types
numériques qu'on utilisera dans ce cours seront :

-   **int** : nombres entiers

-   **float** : nombres réels

-   **complex** : nombres complexes

On utilisera également les types suivants :

-   **str** : chaîne de caractère

-   **list** : pour une liste

-   **bool** : pour un booléen qui peut prendre une valeur 'True' ou
    'False'

La commande **type()** permet de connaître le type d'une variable.

Tout nombre écrit sans point est de type entier. Avec un point, c'est un
nombre de type **float** : **10** et **10.** ne sont pas de même type.

````python
a=10
print(a)
````

````python
type(a)
````

Tout nombre écrit sans point est de type entier. Avec un point, c'est un
nombre de type `float` : `10` et `10.` ne sont pas de même type:

````python
a=10
print(type(a))
````

````python
b=10.
print(type(b))
````

La notation pour les nombres complexe est :

partie réelle + partie imaginaire `j`

Exemple :

````python
c=3+2j 
print(c)
````

ou alternativement :

````python
d=complex(3, 2)  
print(d)
````

Il est possible de changer le type d'une variable numérique en utilisant
une des commandes suivants: **int()**, **float()**, **complex()** :

````python
c=complex(a)
print(c, type(c))
````

````python
a=float(a)
print(a, type(a))
````

## Opérateurs arithmétiques

Les opérateurs sont des symboles spéciaux en Python qui effectuent des
calculs arithmétiques ou logiques. En Python, les opérateurs
arithmétiques sont :

-   `+` : addition

-   `-` : soustraction

-   `*` : multiplication

-   `/` : division

-   `//` : quotient de la division euclidienne (retourne un entier)

-   `%` : reste de la division euclidienne (retourne un entier)

-   `**` : puissance

Ces symboles ont un sens mathématique avec des variables numériques ; 
avec les autres ça dépend. Ils suivent les règles de distributivité et
d'associativité habituelles, avec les parenthèses `(...)`.

Il existe aussi des opérateurs logiques (séance 2), d'affectation
(séance 3), binaires, etc.

**Exemples :**

````python
x = 1
y = 2
a = 3
b = 4
c = 5
print("a*x + b =", a*x + b)
print("3*x**2 - 4*y**2 =", 3* x**2 - 4*y**2) 
print("3*(x+y) / (a+b) =", 3*(x+y) / (a+b))
print("c // y =", c // y)
````

:::{tip}
Les espaces aident à rendre le code lisible et il est ainsi plus facile à débugguer!
:::

Les opérateurs arithmétiques n'ont pas toujours de sens pour les
variables non numériques:

````python
texte='bonjour cher ami'
nouveau_texte = texte - "ami"
print(nouveau_texte)
````

ou entre types différents. Exemple :

````python
a='3'
b=2
print(a+b)
````

````python
print(int(a)+b)
````

## Import

La fonction `import` permet d'importer des librairies, qui sont des
extensions au python de base. Par exemple la librairie `math` contient
les principales fonctions mathématiques. Les méthodes contenues dans la
librairie sont accessibles via un point :

````python
import numpy as np 
# numpy est la librairie de base pour les mathematiques, ici renommée en "np"
# on oubliera la librairie math (pour ceux qui la connaissent)

b = 5
print(np.sqrt(b))
````

Une fois importée dans une cellule, **la librairie est visible par
toutes les autres**, comme vos variables : la mémoire de votre notebook
est partagée entre toutes les cellules.

````python
print(np.sin(np.pi), np.cos(np.pi))
````

On peut aussi n'importer que quelques méthodes d'une librairie, avec la
syntaxe `from... import...` :

````python
from numpy import cos, sin, pi

print(sin(pi), cos(pi))
````

Notez que dans ce cas, `sin` et `cos` sont accessibles sans mettre
`np.` devant ! 

:::{warning}
Attention alors à ne pas définir une variable qui
s'appelle `sin`, sinon il faudra la réimporter\...
:::
