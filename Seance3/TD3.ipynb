## **Exercice 11 :** Dessins en ASCII
Vous souhaitez imprimer à l'écran un carré vide de taille $n$ x $n$ ou un triangle bas de même taille. Le programme ci-dessous imprime à l'écran un carré plein de côté $n=5$. En faisant attention à l'indentation pour chaque bloc d'instructions, copiez ce programme et exécutez le :
```python
#Change la taille du carre
n = 5

#Imprime a l'ecran
for i in range(n):
    for j in range(n):
        print('*',end=' ')
    print('')
```

* Supprimer ``,end=' '`` après ``print('*' ``. Que se passe-t-il et pourquoi ?
